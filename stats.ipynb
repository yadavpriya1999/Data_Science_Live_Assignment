Skewness is a measure of the asymmetry of the probability distribution of a real-valued random variable about its mean. It indicates whether the data points are skewed to the left (negative skewness) or to the right (positive skewness) of the mean.

Types of Skewness

1.Positive Skewness (Right Skewness):


In a positively skewed distribution, the tail on the right side of the distribution is longer or fatter than the left side.
Most of the data points are concentrated on the left side, with fewer larger values on the right side.
Mean > Median > Mode

2.Negative Skewness (Left Skewness):


In a negatively skewed distribution, the tail on the left side of the distribution is longer or fatter than the right side.
Most of the data points are concentrated on the right side, with fewer smaller values on the left side.
Mean < Median < Mode

3.Zero Skewness (Symmetrical Distribution):


In a symmetrical distribution, the left and right sides of the distribution are mirror images of each other.
The mean, median, and mode are all equal or nearly equal.

Correlation is a statistical measure that describes the strength and direction of a relationship between two variables. It quantifies how one variable changes in relation to another.

Types of Correlation


1.Positive Correlation:

When one variable increases, the other variable also increases.
Values range from 0 to +1.
Example: Height and weight typically have a positive correlation; as height increases, weight tends to increase.

2.Negative Correlation:

When one variable increases, the other variable decreases.
Values range from 0 to -1.
Example: The amount of time spent studying and the number of errors on a test might have a negative correlation; as study time increases, errors tend to decrease.

3.No Correlation:

No relationship between the variables.
The correlation value is around 0.
Example: The number of books read and the number of movies watched may have no correlation.

```python
import numpy as np

# Data
accountancy = np.array([45, 70, 65, 30, 90, 40, 50, 75, 85, 60])
statistics = np.array([35, 90, 70, 40, 95, 40, 60, 80, 80, 50])

# Number of students
n = len(accountancy)

# Calculate sums
sum_x = np.sum(accountancy)
sum_y = np.sum(statistics)
sum_x2 = np.sum(accountancy**2)
sum_y2 = np.sum(statistics**2)
sum_xy = np.sum(accountancy * statistics)

# Calculate the correlation coefficient (r)
numerator = n * sum_xy - sum_x * sum_y
denominator = np.sqrt((n * sum_x2 - sum_x**2) * (n * sum_y2 - sum_y**2))
r = numerator / denominator

r
```

```python
#28.What is Normal Distribution? What are the four Assumptions of Normal Distribution? Explainin detail.
```

The Normal Distribution, also known as the Gaussian distribution, is a continuous probability distribution that is symmetrical around its mean, showing that data near the mean are more frequent in occurrence than data far from the mean. In graph form, it appears as the classic bell-shaped curve.

Four Assumptions of Normal Distribution

1.Randomness:

The data points should be collected randomly from the population of interest.

2.Independence:

The observations should be independent of each other.

3.Identical Distribution:

All observations should come from the same underlying distribution.

4.Continuous Data:

The data should be continuous and measured at an interval or ratio scale.


```python
import numpy as np
import matplotlib.pyplot as plt

# Generate random data
mean = 0
std_dev = 1
data = np.random.normal(mean, std_dev, 1000)

# Plot the histogram
plt.hist(data, bins=30, density=True, alpha=0.6, color='b')

# Plot the normal distribution curve
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = np.exp(-((x - mean) ** 2) / (2 * std_dev ** 2)) / (np.sqrt(2 * np.pi) * std_dev)
plt.plot(x, p, 'k', linewidth=2)
plt.title("Normal Distribution")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

Characteristics or Properties of the Normal Distribution Curve

1.Symmetry:

The normal distribution curve is symmetrical around the mean.
This means that the left and right halves of the curve are mirror images of each other.

2.Mean, Median, and Mode:

In a normal distribution, the mean, median, and mode are all equal and located at the center of the distribution.
This central point is the highest peak of the bell-shaped curve.

3.Bell-Shaped Curve:

The shape of the normal distribution curve is bell-shaped, indicating that most of the data points are concentrated around the mean.
The tails of the curve extend indefinitely but never touch the horizontal axis, approaching zero asymptotically.

3.Asymptotic Nature:

The tails of the normal distribution curve approach the horizontal axis asymptotically, meaning they get closer and closer to the axis but never actually reach it.
This implies that there is always a small probability of extreme values occurring.

4.Empirical Rule (68-95-99.7 Rule):

Approximately 68% of the data falls within one standard deviation of the mean.
About 95% of the data falls within two standard deviations of the mean.
Nearly 99.7% of the data falls within three standard deviations of the mean.

5.Unimodal:

The normal distribution has a single peak (unimodal), indicating that there is only one mode, or most frequent value, in the dataset.

6.Total Area Under the Curve:

The total area under the normal distribution curve is equal to 1, representing the total probability of all possible outcomes.

7.Inflection Points:

The points where the curve changes concavity (from concave up to concave down or vice versa) are called inflection points.
In a normal distribution, the inflection points are located at one standard deviation away from the mean on either side.

```python
#31. The mean of a distributionis 60 with a standard deviation of1 0.Assuming that the distribution Is normal, what percentage of Items be (I)between 60 and 72, (II) between 50 and 60, (Ill) beyond 72 and (Iv) between 70 and 80?
```

```python
import scipy.stats as stats

# Given parameters
mean = 60
std_dev = 10

# Calculate the Z-scores
z_60 = (60 - mean) / std_dev
z_72 = (72 - mean) / std_dev
z_50 = (50 - mean) / std_dev
z_70 = (70 - mean) / std_dev
z_80 = (80 - mean) / std_dev

# Calculate the percentages using the cumulative distribution function (CDF)
# Percentage between 60 and 72
percentage_60_72 = stats.norm.cdf(z_72) - stats.norm.cdf(z_60)

# Percentage between 50 and 60
percentage_50_60 = stats.norm.cdf(z_60) - stats.norm.cdf(z_50)

# Percentage beyond 72
percentage_beyond_72 = 1 - stats.norm.cdf(z_72)

# Percentage between 70 and 80
percentage_70_80 = stats.norm.cdf(z_80) - stats.norm.cdf(z_70)

# Convert to percentages
percentage_60_72 *= 100
percentage_50_60 *= 100
percentage_beyond_72 *= 100
percentage_70_80 *= 100

# Print the results
print(f"Percentage of items between 60 and 72: {percentage_60_72:.2f}%")
print(f"Percentage of items between 50 and 60: {percentage_50_60:.2f}%")
print(f"Percentage of items beyond 72: {percentage_beyond_72:.2f}%")
print(f"Percentage of items between 70 and 80: {percentage_70_80:.2f}%")
```

a. Statistical Hypothesis

A statistical hypothesis is a statement or an assumption about a population parameter. It is a conjecture or prediction that can be tested through statistical methods. There are two types of hypotheses:

1.Null Hypothesis (Hâ‚€):

This is a statement of no effect or no difference. It assumes that any observed difference or effect is due to chance or random variability.
Example: Hâ‚€: Î¼ = 50 (The mean of the population is 50)

2.Alternative Hypothesis (Hâ‚ or Ha):

This is a statement that contradicts the null hypothesis. It suggests that there is an effect or a difference.
Example: Hâ‚: Î¼ â‰  50 (The mean of the population is not 50)

Errors in Hypothesis Testing

In hypothesis testing, there are two types of errors that can occur:

Type I Error (Î±):

Occurs when the null hypothesis is rejected when it is actually true.
This is also known as a "false positive."
The probability of making a Type I error is denoted by alpha (Î±) and is also known as the significance level of the test.
Example: Concluding that a new drug is effective when it is actually not.

Type II Error (Î²):

Occurs when the null hypothesis is not rejected when it is actually false.
This is also known as a "false negative."
The probability of making a Type II error is denoted by beta (Î²).
Example: Concluding that a new drug is not effective when it actually is.

```python
#35.A random sample of size 25 froma population gives the sample standard derivationto be 9.0. Test the hypothesis that the population standard derivation is 10.5.Hint(Usechi-square distribution).
```

```python
import scipy.stats as stats

# Given parameters
n = 25
s = 9.0
sigma_0 = 10.5
alpha = 0.05

# Calculate the test statistic
test_statistic = (n - 1) * (s ** 2) / (sigma_0 ** 2)

# Degrees of freedom
df = n - 1

# Critical values for two-tailed test
critical_value_lower = stats.chi2.ppf(alpha / 2, df)
critical_value_upper = stats.chi2.ppf(1 - alpha / 2, df)

# Print the results
print(f"Test Statistic: {test_statistic:.2f}")
print(f"Critical Value Lower: {critical_value_lower:.2f}")
print(f"Critical Value Upper: {critical_value_upper:.2f}")

# Determine if we reject or fail to reject the null hypothesis
if test_statistic < critical_value_lower or test_statistic > critical_value_upper:
    print("Reject the null hypothesis. The population standard deviation is significantly different from 10.5.")
else:
    print("Fail to reject the null hypothesis. There is not enough evidence to say the population standard deviation is different from 10.5.")
```

```python
import scipy.stats as stats

# Given data
grades = ['A', 'B', 'C', 'D', 'E']
observed_frequencies = [15, 17, 30, 22, 16]
total_students = 100
num_grades = len(grades)

# Calculate the expected frequencies
expected_frequency = total_students / num_grades
expected_frequencies = [expected_frequency] * num_grades

# Calculate the Chi-square test statistic
chi2_statistic, p_value = stats.chisquare(f_obs=observed_frequencies, f_exp=expected_frequencies)

# Degrees of freedom
df = num_grades - 1

# Critical value for alpha = 0.05
alpha = 0.05
critical_value = stats.chi2.ppf(1 - alpha, df)

# Print the results
print(f"Chi-square Statistic: {chi2_statistic:.2f}")
print(f"Critical Value at alpha = 0.05: {critical_value:.2f}")
print(f"P-value: {p_value:.4f}")

# Determine if we reject or fail to reject the null hypothesis
if chi2_statistic > critical_value:
    print("Reject the null hypothesis. The distribution of grades is not uniform.")
else:
    print("Fail to reject the null hypothesis. There is not enough evidence to say the distribution of grades is not uniform.")
```
