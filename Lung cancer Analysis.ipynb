```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Data set\dataset.csv')
```

```python
df.head()
```

```python
# Display basic information about the dataset
print("Dataset Information:")
print(df.info())
```

Insight: No any null data

```python
#data summary
print(df.describe())
```

```python
# Check for missing values
print("\nMissing Values:")
print(df.isnull().sum())
```

```python
# Data type conversion if necessary (e.g., convert GENDER and LUNG_CANCER to categorical)
df['GENDER'] = df['GENDER'].astype('category')
df['LUNG_CANCER'] = df['LUNG_CANCER'].astype('category')
```

```python
# Distribution of features
df.hist(bins=30, figsize=(20, 15))
plt.suptitle('Histograms of Features')
plt.show()
```

Insight:Histograms will help you understand the distribution of each numerical feature.

```python
# Box plots for detecting outliers
plt.figure(figsize=(12, 8))
sns.boxplot(data=df.drop(columns=['GENDER', 'LUNG_CANCER']))
plt.xticks(rotation=90)
plt.title('Box Plots of Numerical Features')
plt.show()
```

Insight:Box plots can be used to identify outliers in the numerical features.

```python
# Heatmap of correlations
plt.figure(figsize=(12, 8))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

Insight:A correlation heatmap will help you understand the relationships between different numerical features.
